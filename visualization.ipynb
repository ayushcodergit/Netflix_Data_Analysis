```python
#Visualizing Netflix data using charts and plots

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv("cleaned_netflix_data.csv")

# Set plot style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

#Movies vs TV Shows
sns.countplot(data=df, x='type', palette='Set2')
plt.title("Content Type Distribution (Movies vs TV Shows)")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()

#Content added per year
df['year_added'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Titles Added to Netflix Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()

#Top 10 Countries
top_countries = df['country'].str.split(',').explode().str.strip().value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='coolwarm')
plt.title("Top 10 Content-Producing Countries")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()

#Rating distribution
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index, palette='magma')
plt.title("Ratings Distribution")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()

#Top 10 Genres
top_genres = df['listed_in'].str.split(',').explode().str.strip().value_counts().head(10)
top_genres.plot(kind='bar', color='teal')
plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```
