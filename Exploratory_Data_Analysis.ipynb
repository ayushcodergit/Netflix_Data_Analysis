```python
#Exploratory Data Analysis - uncover trends and patterns


import pandas as pd

# Load cleaned dataset
df = pd.read_csv("cleaned_netflix_data.csv")

#Content Type Count
content_counts = df['type'].value_counts()
print("\nContent Type Distribution:\n", content_counts)

#Content Added Per Year
year_wise = df['year_added'].value_counts().sort_index()
print("\nContent Added per Year:\n", year_wise)

#Top 10 Countries by Content Count
top_countries = df['country'].str.split(',').explode().str.strip().value_counts().head(10)
print("\nTop 10 Countries:\n", top_countries)

#Rating Distribution
rating_counts = df['rating'].value_counts()
print("\nRating Distribution:\n", rating_counts)

#Top 10 Genres
top_genres = df['listed_in'].str.split(',').explode().str.strip().value_counts().head(10)
print("\nTop 10 Genres:\n", top_genres)  

# Save EDA output to a text file (optional)
with open("eda_summary.txt", "w") as file:
    file.write("Content Type Distribution:\n" + content_counts.to_string() + "\n\n")
    file.write("Content Added per Year:\n" + year_wise.to_string() + "\n\n")
    file.write("Top 10 Countries:\n" + top_countries.to_string() + "\n\n")
    file.write("Rating Distribution:\n" + rating_counts.to_string() + "\n\n")
    file.write("Top 10 Genres:\n" + top_genres.to_string() + "\n")

print("\nEDA summary saved as 'eda_summary.txt'")
```
